## 4.- Emphasis on Strings

A string is a set of characters in a specific order. They can create a word, a sentence, or represent things such as DNA or protein sequences. The mnethods of a string obhect make it really easy to work with sequences in python.



```python
#### EXAMPLE:

sequence1 = 'AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC'


### run the line of code to find all the methods that a sting type has.

dir(sequence1)
```

```python
### Look at the count method attached to a string,
G = sequence1.count('G')
C = sequence1.count('C')

print(G)
print(C)
```

```python
## Content of G and C in sequence

def gc_content(sequence1):
  G = sequence1.count('G')
  C = sequence1.count('C')
  gc_content = G + C
  return gc_content

gc_content(sequence1)
```

```python
#### SIMPLE EXERCISE:
## Write a function that takes a single string as input and calculates the GC content of it
## GC content is the percentage (or fraction) of a sequence that it is G or C.
# def gc_calculations(seq):



def gc_calculations(seq):
  G = seq.count('G')
  C = seq.count('C')
  gc_percent = (G + C) / len(seq) *100
  return gc_percent

gc_calculations(sequence1)
```
